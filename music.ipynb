```python
import pandas as pd
import warnings
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
# from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score
import joblib
```

```python
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
music_df = pd.read_csv('music.csv')
music_df
```

```python
data = music_df.drop(columns=['genre'])
genre = music_df['genre']
# data_train, data_test, genre_train, genre_test = train_test_split(
#     data, genre, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(data, genre)
joblib.dump(model, 'musicRecommender.bin')

# predictions = model.predict(data_test)
# accuracy = accuracy_score(genre_test, predictions)
# print(f"Accuray is {accuracy * 100}%")
```

```python
tree.export_graphviz(model, out_file='musicRecommender.dot',
                     feature_names=['age', 'gender'], class_names=sorted(genre.unique()),label='all',rounded=True,filled=True)
```
